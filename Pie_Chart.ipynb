```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.2f%%', shadow=True, explode=[0,0.2,0,0.2,0], startangle=45)
plt.show()
plt.savefig("pie_chart.png", bbox_inches="tight", pad_inches=1)
plt.savefig("pie_chart.pdf", bbox_inches="tight", pad_inches=1)
```
