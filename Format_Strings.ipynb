```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7] # week days
y = [50,51,52,48,47,49,46]  # temperature
```

```python
plt.plot(x,y, 'g+-.')   # g=green, +=point, -.=style
```

```python
plt.plot(x,y,'rd-')
```

```python
plt.plot(x,y, color='red', marker='D', linestyle='-', markersize='10')
```

```python
plt.plot(x,y, color='green', alpha=0.4)   # alpha used for transperacy scale between 0-1
```
